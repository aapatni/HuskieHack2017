```python
import pandas as pd
import numpy as np
from datetime import datetime
import random
dataFrame = pd.read_csv("disaster.csv")
#dataFrame2 = pd.read_csv('disaster2.csv')


dataFrame = dataFrame.loc[:,["State","Disaster Type", "Start Date"]]

#Create the State List and Type List
stateDict = {}
stateList = []
typeDict = {}
typeList = []
for i in dataFrame["State"]:
    stateDict[i] = i
#for i in dataFrame2['state']:
 #   stateDict[i] = i
for key, value in stateDict.items():
    stateList.append(key)


for i in dataFrame["Disaster Type"]:
    typeDict[i.lower()] = i
#for i in dataFrame2['incidentType']:
 #   typeDict[i.lower()] = i
for key, value in typeDict.items():
    typeList.append(key)
###End of creating State List and Type list


dataFrame['State'] = dataFrame['State'].apply(lambda x: stateList.index(str(x)))
dataFrame['Disaster Type'] = dataFrame['Disaster Type'].apply(lambda x: typeList.index(str(x).lower()))
dataFrame['Start Date'] = dataFrame["Start Date"].apply(lambda x: int(str(x).partition("/")[0]))


    
def checkaroo(event,x):
    if event == x:
        return 1
    return 0
for event in typeList:
    dataFrame[event] = dataFrame['Disaster Type'].apply(lambda x: checkaroo(event,typeList[x]))

dataFrame["randomize"]= dataFrame['Disaster Type'].apply(lambda x: random.randint(0,40000))
dataFrame = dataFrame.sort_values(by=["randomize"])
dataFrame
#dataFrame2["state"] = dataFrame2['state'].apply(lambda x: stateList.index(str(x)))
#dataFrame2['incidentType'] = dataFrame2['incidentType'].apply(lambda x: typeList.index(str(x).lower()))
#dataFrame2['declarationDate'] = dataFrame2['declarationDate'].apply(lambda x: int(datetime.strptime(str(x)[:10],'%Y-%m-%d').month))
```

```python

x_1 = np.array([dataFrame["State"].values, dataFrame["Start Date"].values], dtype=np.float32).transpose()
#x_2 = np.array([dataFrame2["state"].values, dataFrame2["declarationDate"].values]).transpose()
y_1 = []
for disaster in typeList:
    y_1.append( np.array([dataFrame[disaster].values], dtype=np.float32).transpose()) 

#y_2 = np.array([dataFrame2["incidentType"].values]).transpose()
#X =  np.concatenate((x_1,x_2),axis=0)
#Y =  np.concatenate((y_1,y_2),axis=0)
X= x_1
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn import tree
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
#clf_tree = tree.DecisionTreeClassifier()

#clf = MLPClassifier()
#clfGrad = GradientBoostingClassifier()
binaryClfs = []
for data in range(len(y_1)):
    Y = y_1[data].ravel()
    #clfBinary = tree.DecisionTreeClassifier()
    #clfBinary.fit(X,Y)
    clfBinary = LogisticRegression(max_iter = 5000000, solver="sag",n_jobs=5)
    clfBinary.fit(X,Y)
    print("One done")
    binaryClfs.append(clfBinary)
#clf_tree.fit(X,Y.ravel())
#clf.fit(X,Y.ravel())
#clfGrad.fit(X,Y.ravel())
```

```python
month = int(input("What is the month number: "))
state = stateList.index(str(input("State Abbreviation: ")))
probabilities = []
for num in range(len(binaryClfs)):
    clf = binaryClfs[num]
    disaster = typeList[num]
    predic = clf.predict([[month,state]])
    deci = clf.decision_function([[month,state]])
    prob = clf.predict_proba([[month,state]])
    thing = [disaster,predic, prob]
    probabilities.append(thing)

for i in probabilities:
    print (i)
    
```

```python
def testVal(clf, X_test, Y_test):
    if (clf.predict(X_test)==Y_test[0]):
        return 1
    return 0

counterBinary = []
for clf in binaryClfs:
    counterBinary.append(0)

    
i=0
while True:
    try:
        #print(i,stateList[X[i][0]], X[i][1], typeList[Y[i][0]])
        for x in range(len(binaryClfs)):
            current = binaryClfs[x]
            counterBinary[x]+= testVal(current,x_1[i].reshape(1,-1),y_1[x][i])
            print(current.score(x_1[i].reshape(1,-1),y_1[x][i]))
        i+=1
    except Exception as e:
        #print(counter/float(i-1))
        print(e)
        %tb
        print(counterBinary)
        #print(counterGrad)
        break
        
        
    
```

```python
#binaryClfs - is a list that contains all the classifiers
import pickle

filename = 'finalized_model'

for i in range(len(binaryClfs)):
    pickle.dump(binaryClfs[i], open(filename+str(i)+".sav", 'wb'))

pickle.dump(stateList, open("statelist.txt","wb"))
pickle.dump(typeList, open("typelist.txt","wb"))
```

```python
def main(state, month):
    moreClfs = []
    stateList1 = pickle.load(open("statelist.txt","rb"))
    typeList1 = pickle.load(open("typelist.txt","rb"))
    state = int(stateList1.index(state))
    for i in range(len(binaryClfs)):
        loaded_model = pickle.load(open(filename+str(i)+".sav", 'rb'))
        moreClfs.append(loaded_model)

    for clf in binaryClfs:
        disaster = typeList1[binaryClfs.index(clf)]
        predic = clf.predict([[month,state]])
        prob = clf.predict_proba([[month,state]])

        print([disaster,predic,prob])
        
main("AL",5)
```
